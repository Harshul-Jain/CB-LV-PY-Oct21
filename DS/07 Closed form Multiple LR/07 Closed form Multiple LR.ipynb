```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

## Generate Data

```python
X, y, coef = make_regression(n_samples=100, n_features=1, bias=10, noise=10, coef=True)
print(X.shape, y.shape)
print(coef)
```

```python
plt.scatter(X, y)
# plt.scatter(0,10, c='red')
plt.show()
```

```python
def train(X, Y):
    X = X.copy()
    ones_column = np.ones((len(X),1))
    X = np.concatenate([ones_column, X], axis=1)
    W = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
    return W
    
```

```python
w = train(X,y)
print(w)
```

```python
c, m = w
print(c,m)
```

```python
x = np.linspace(X.min(), X.max(), 100)
y_hat = m*x + c
plt.scatter(X, y, label='data')
plt.plot(x, y_hat, c='red', label='Prediction line') # best fit line
plt.legend()
plt.show()
```
