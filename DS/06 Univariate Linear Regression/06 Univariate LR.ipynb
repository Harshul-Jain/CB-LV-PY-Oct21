```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dirpath = "../data/knn/regression/"
X = np.load(dirpath+"X_data.npy")
y = np.load(dirpath+"Y_data.npy")
print(X.shape, y.shape)
```

```python
plt.scatter(X, y)
plt.xlabel("Feature")
plt.ylabel("Label/ Output")
plt.title("Label vs Feature")
plt.show()
```

## Optimal parameters

```python
m = ((X[:,0]*y).mean() - X[:,0].mean()*y.mean())/((X[:,0]**2).mean()- (X[:,0].mean()**2)) # optimal value of m
c = y.mean() - m*X[:,0].mean()
print(m, c)
```

## Plot the Prediction line

```python
x = np.linspace(X[:,0].min(), X[:,0].max(), 100)
print(x)
```

```python
y_hat = m*x + c # prediction
plt.scatter(X, y, label='Training points')
plt.plot(x, y_hat, c='red', label="Best fit line")
plt.legend()
plt.xlabel("Feature")
plt.ylabel("Label/ Output")
plt.title("Label vs Feature")
plt.show()
```
